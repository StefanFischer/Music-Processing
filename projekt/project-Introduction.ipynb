# Project: Chord Recognition


## Problem Statement 

If you want to play to your favourite songs but you are unlucky for having a very specific music taste, there is sometimes the problem that there are no chords for the songs to play along. Therefore we want to develop a simple chord recognition application to help you playing along to your favourite songs!

Sometimes chords are not available for a specific song:
* song is brand new
* band is under the radar
* it is originally not performed on guitar/piano


## Idea

Using music processing analysis we want to build a simple yet powerful tool to extract the chords of an audio file and present it to the user in a way the user can play along with the audio recording. Therefore we want to apply a chord recognition algorithm with a rather musical approach.

### Strategy

Musical Approach: 
1. Use HPS to process harmonics and percussion signal seperately
2. Seperate the music signal into beats with the beat tracking algorithm into musically meaningful junks
3. Recognize chord of current beat

### User Inteface

1. Playback of music recording
2. Visualization of recognized chords to be able to play along to the music recording
3. Additional transpose feature to adjust chords (guitar capodaster/piano transpose)

### Beat Tracking

As our application should be mainly for modern popular music, we assume that the underlying tempo of the songs is constant. Therefore we will use Beat Tracking by Dynamic Programming. Furthermore, our input signal will be the percussive component of the signal generated by the HPS.


### Chord Recognition

As input signal use the harmonic component computed by the HPS. Besides that, we want to use a window of windowsize of one beat, computed by the beat tracking algorithm.
